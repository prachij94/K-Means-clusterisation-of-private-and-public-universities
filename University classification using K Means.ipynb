We will attempt to use KMeans Clustering to cluster some Universities into to two groups, **Private** and **Public**.

K Means Clustering is an unsupervised learning algorithm that tries to cluster data based on their similarity. Unsupervised learning means that there is no outcome to be predicted, and the algorithm just tries to find patterns in the data. In k means clustering, we have the specify the number of clusters we want the data to be grouped into i.e. 2 in this case. 

The Data has 777 observations on the following 18 variables.
* Private: A factor with levels No and Yes indicating private or public university
* Apps: Number of applications received
* Accept: Number of applications accepted
* Enroll: Number of new students enrolled
* Top10perc: Pct. new students from top 10% of H.S. class
* Top25perc: Pct. new students from top 25% of H.S. class
* F.Undergrad: Number of fulltime undergraduates
* P.Undergrad: Number of parttime undergraduates
* Outstate: Out-of-state tuition
* Room.Board: Room and board costs
* Books: Estimated book costs
* Personal: Estimated personal spending
* PhD: Pct. of faculty with Ph.D.â€™s
* Terminal: Pct. of faculty with terminal degree
* S.F.Ratio: Student/faculty ratio
* perc.alumni: Pct. alumni who donate
* Expend: Instructional expenditure per student
* Grad.Rate: Graduation rate

## K Means Cluster Creation


```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=2)
```

```python
km.fit(data.drop('Private',axis=1))
```

```python
km.cluster_centers_
```

```python
km.labels_
```

## Evaluation

There is no perfect way to evaluate clustering if you don't have the labels.

```python
data['Cluster']=data['Private'].replace({'Yes':1,'No':0})
```

```python
data
```

Run PCA on the data and reduce the dimensions to 2 dimensions

```python
from sklearn.decomposition import PCA
```

```python

reduced_data = PCA(n_components=2).fit_transform(data.drop('Private',axis=1))
pca_df = pd.DataFrame(reduced_data,columns=['pca1','pca2'])
```

```python
sns.set_style('darkgrid')
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(16,10))
ax1.set_title('K Means')
ax1.scatter(pca_df['pca1'],pca_df['pca2'],c=km.labels_,cmap='magma',alpha=0.8)
ax2.set_title("Original")
ax2.scatter(pca_df['pca1'],pca_df['pca2'],c=data['Cluster'],cmap='magma',alpha=0.8)
```
